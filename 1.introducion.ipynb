### Que es python?[*](https://www.python.org/).
 
Es un lenguaje de programación interpretado con una sintaxis fácil de aprender orientado a objeto es también un lenguaje de script puedes desarrollar programas para varios sistemas operativos  (linux, windows, macOS etc).
 
### que significa python sea interpretado?[*](https://en.wikipedia.org/wiki/Interpreter_(computing)#Compilers_versus_interpreters).
 
a diferencia de otros lenguajes de programación que es necesario compilar el código para luego ser traducido el código de lenguaje humano a lenguaje máquina (código binario 0 y 1) en python esta compilación se realiza de manera interna el compilador de python va realizado la traducción durante el tiempo de ejecución

la sintaxis es lo más parecido a la ortografía de un idioma, python su sintaxis es simple y fácil de aprender.

> 1
> >>> 3
>3

1. variables:
 
una variable en los lenguaje de programación son lugares (esto lugare son llamdo como  ***espacios en memoria***) donde se guardan información que irás utilizando en durante el programa pero esta información puede ir cambiando o cambiando. Son como cajas donde vas guardando objetos puede que los que guarde en esa caja cambio lo que guarde en ella
 
en python una variable guridad diferentes infomacion estas serían:
* texto: los tipos texto se identifica con comillas simples ('') o dobles ("")
* numeros
* numeros con comas
* boleanos

3. texto
los texto se representa con comillas simples ('') o dobles ("")
>```
> >>> "the cat is..."
> 'the cat is...'
> >>> 'Don\'t is.. '
> "Don't is.. "
> >>> "the 'cat' is not man"
> "the 'cat' is not man"
> >>> "the "cat" is not man"
> File "<stdin>", line 1
>    "the "cat" is not man"
>          ^^^
> SyntaxError: invalid syntax
>```
 
si colocas ``"the "cat" is not man"`` esto da error por que reconoce como 2 testo diferentes ``"the"`` son texto  por estar dentro de comillas dobles``"is not man"`` pero cat está fuera de las comillas dobles al estar fuera de las comillas dobles el intérprete lo reconoce como si fuera una variable una forma de resolverlo sería utilizando una barra invertida \  
 
>```
> >>> "the \"cat\" is not man"
> 'the "cat" is not man'
>```
estás barra invertidas en un texto se llaman escapes nos sirven para escapar un comillas en un texto sino también realizar otras cosas como
 
* \n nueva linea
* \t un tabulation
 
>```
> >>> "the is a line\n is a new line"
> "the is a line\n is a new line"
> >>> print("the is a line\nis a new line")
> the is a line
> is a new line
>```
 
al escribir en el propio intérprete el texto te darás cuenta de que tanto al final como al inicio aparecerán comillas dobles o simples y al colocar un salto de línea sólo aparece en el texto y no realizar un salto de línea.
 
esto se debe a que el intérprete le los texto como texto plano sin formatos al añadirle un salto de línea le está añadiendo formato al texto
 
una forma de resolver ese problema es utilizar una función definida llamada ``print()``
